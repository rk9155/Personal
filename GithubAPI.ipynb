```python
# Fetch the cuisine_id of of cuisine BBQ using Zomato API.
import requests
from requests.auth import HTTPBasicAuth
import json

user_key = '47fde626b66911190ddd56c9e0d63f9b'
url = "https://developers.zomato.com/api/v2.1"
params = {'city_id' : 1}
header = {'Accept': 'application/json','user-key': user_key}
response = requests.get(url+'/cuisines',headers=header,params = params)
cuisine_id = 0
for i in response.json().get('cuisines'):
    if(i.get('cuisine').get('cuisine_name')== 'BBQ'):
        cuisine_id = i.get('cuisine').get('cuisine_id')
print(cuisine_id)
```
