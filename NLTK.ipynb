## Movie Reviews Project

```python
from nltk.corpus import movie_reviews
nltk.download('movie_reviews')
  
```

```python
movie_reviews.fileids()
```

```python
movie_reviews.words(movie_reviews.fileids()[6])
```

```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))

documents[0:4]
    
```

```python
import random
random.shuffle(documents)
documents[0:4]
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)

from nltk.corpus import wordnet
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
def clean_reviews(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_words = lemmatizer.lemmatize(w, pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
            
    return output_words
```

```python
documents = [(clean_reviews(documents),category)for documents,category in documents]
```

```python
documents[0]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
common
```

```python
features = [i[0] for i in common]
features
```

```python
def get_features_dict(words):
    current_features = {}
    word_set = set(words)
    for w in features:
        current_features[w] = w in word_set
    return current_features
```

```python
output = get_features_dict(documents[0][0])
output
```

```python
training_data = [(get_features_dict(doc),category) for doc,category in training_documents]
```

```python
testing_data = [(get_features_dict(doc),category) for doc,category in testing_documents]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
classifier.show_most_informative_features(15)
```
