```python
def step_gradient(points,learning_rate,m):
    M=len(points)
    m_slope=list([0 for i in range(len(points[0])-1)])
    for i in range(M):
        y=points[i,len(points[0])-2]
        x_total=0;
        q=0
        for k in range(len(points[0])):
            if(k==len(points[0])-2):
                continue
            x_total+=m[q]*points[i,k]
            q=q+1
        l=0   
        for j in range(len(points[0])):
            if (j==len(points[0])-2):
                continue
            m_slope[l]+=(-2/M)*((y-x_total)*points[i,j])
            l=l+1
    newm=list([0 for j in range(len(points[0])-1)])
    a=0
    for i in range(len(m)):
        newm[i]=m[i]-learning_rate*m_slope[i]
    return newm
```

```python
def gd(points,learning_rate,iteration):
    m=list([0 for j in range(len(points[0])-1)])
    for i in range(iteration):
        m=step_gradient(points,learning_rate,m)
        print(i," cost : ",cost(points,m))
    return m
```

```python
def cost(points,m):
    total_cost=0
    M=len(points)
    total_cost=0
    for i in range(M):
        y=points[i,len(points[0])-2]
        x_total=0
        q=0
        for k in range(len(points[0])):
            if(k==len(points[0])-2):
                continue
            x_total+=m[q]*points[i,k]
            q=q+1
        total_cost+=(1/M)*((y-x_total)**2)    
    return total_cost
```
