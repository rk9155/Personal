```python
# Count the total number of cities of country AU 
# from this URL : https://api.openaq.org/v1/cities/
import json
import requests

city = requests.get(' https://api.openaq.org/v1/cities/')
city_info = json.loads(city.text)
city_res = city_info['results']
count = 0
for i in city_res:
    if(i['country']=='AU'):
        count+=1
print(count)

```

```python
# Print first 5 cities from country “BE”. Fetch the result from given URL.
```

```python
import json
import requests

city = requests.get('https://api.openaq.org/v1/cities/',params={"country":"BE","limit":5})
city_info = json.loads(city.text)
city_res = city_info['results']
for i in city_res:
    print(i['city'])
```
