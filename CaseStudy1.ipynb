```python
# 1.. Check the trend of investments over the years. To check the trend, find -
# Total number of fundings done in each year.
# Plot a line graph between year and number of fundings. Take year on x-axis and number of fundings on y-axis.
# Print year-wise total number of fundings also. Print years in ascending order.

# year1 TotalFunding1
# year2 TotalFunding2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

data = pd.read_csv("Dataset/startup_funding.csv")
df1 = data.copy()
df1['Date'].replace("12/05.2015","12/05/2015",inplace = True)
df1['Date'].replace("13/04.2015","13/04/2015",inplace = True)
df1['Date'].replace("22/01//2015","22/01/2015",inplace = True)
df1['Date'].replace("15/01.2015","15/01/2015",inplace = True)
date = df1['Date']
year = date.str.split('/',expand = True)[2]
freq = year.value_counts()
x = freq.index
y = freq
plt.plot(x,y)
plt.show()
for i in range(len(x)):
    print(x[i],y[i])
```

```python
# 2..
# Find out which cities are generally chosen for starting a startup.
# Find top 10 Indian cities which have most number of startups ?
# Plot a pie chart and visualise it.
# Print the city name and number of startups in that city also.

# Take city name "Delhi" as "New Delhi".
# Check the case-sensitiveness of cities also. That means - at some place, 
# instead of "Bangalore", "bangalore" is given. 
# Take city name as "Bangalore".
# For few startups multiple locations are given, 
# one Indian and one Foreign. Count those startups in Indian startup also. 
# Indian city name is first.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def handle_city(name):     #added this function to handle multiple city nams.....in case of multiple names...u need to take the first one
    return name.split('/')[0].strip()

data = pd.read_csv("Dataset/startup_funding.csv")
df1 = data.copy()
df1['CityLocation'].dropna(inplace=True)
df1['CityLocation'].replace("Delhi","New Delhi",inplace=True)
df1['CityLocation'].replace("bangalore","Bangalore",inplace=True)

df1['CityLocation'] = df1['CityLocation'].apply(handle_city)

city = df1['CityLocation']
freq = city.value_counts()[0:10]
x = freq.index
y = freq.values
for i in range(len(x)):
    print(x[i],y[i])
plt.pie(y,labels = x,autopct="%.1f%%")
plt.axis("equal")
plt.show()
```

```python
# There are 4 different type of investments. Find out percentage of amount funded for each investment type.
# Plot a pie chart to visualise.
# Print the investment type and percentage of amount funded with 2 decimal places after rounding off.

import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

data = pd.read_csv("Dataset/startup_funding.csv")
df1 = data.copy()
df1['InvestmentType'].replace("SeedFunding","Seed Funding",inplace=True)
df1['InvestmentType'].replace("PrivateEquity","Private Equity",inplace=True)
df1['InvestmentType'].replace("DebtFunding","Debt Funding",inplace=True)
df1['InvestmentType'].replace("Crowd funding","Crowd Funding",inplace=True)
invest = df1['InvestmentType']
freq = invest.value_counts(ascending = True)
x = freq.index
y = freq
plt.pie(y,labels = x,autopct="%.3f%%")
plt.axis("equal")
plt.show()
p = np.true_divide(y,y.sum(axis = 0))*100
# p = p.round(2)
for i in range(len(x)):
    print(x[i],format(p[i],'.2f'))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def modified(amount):                    #added this function to handle the commas in Amounts
    return int(amount.replace(',',''))

data = pd.read_csv("Dataset/startup_funding.csv")
df1 = data.copy()
df1["AmountInUSD"].fillna('0',inplace = True)             #handled nan's
df1["AmountInUSD"] = df1["AmountInUSD"].apply(modified)

df1["InvestmentType"].fillna("",inplace = True)
df1 = df1[df1["InvestmentType"] != ""]


df1['InvestmentType'].replace("SeedFunding","Seed Funding",inplace=True)
df1['InvestmentType'].replace("PrivateEquity","Private Equity",inplace=True)
df1['InvestmentType'].replace("DebtFunding","Debt Funding",inplace=True)
df1['InvestmentType'].replace("Crowd funding","Crowd Funding",inplace=True)

x = df1["InvestmentType"]
y = df1["AmountInUSD"]

d ={}                        #used dictionary to find the fundings in each investment type
for i in x.index:
    d[x[i]] = d.get(x[i],0) + y[i]
         
sum = 0
for i in d:
    sum += d[i]
for i in d:
    print(i,format(d[i]*100/sum,"0.2f"))

plt.pie(d.values(),labels = d.keys())
plt.axis("equal")
plt.show()
```

```python

# 5..
# Which type of companies got more easily funding. To answer this question, find -
# Top 5 industries and percentage of the total amount funded to that industry. (among top 5 only)
# Print the industry name and percentage of the amount funded with 2 decimal place after rounding off.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Dataset/startup_funding.csv")
df1 = data.copy()
df1['IndustryVertical'].replace("eCommerce","Ecommerce",inplace=True)
df1['IndustryVertical'].replace("ecommerce","Ecommerce",inplace=True)
df1['IndustryVertical'].replace("ECommerce","Ecommerce",inplace=True)

# Converting AmountInUsd into numeric

df1['AmountInUSD'] = df1['AmountInUSD'].apply(lambda x :float(str(x).replace(",","")))
df1['AmountInUSD'] = pd.to_numeric(df1['AmountInUSD'])
ind = df1.groupby('IndustryVertical')['AmountInUSD'].sum().sort_values(ascending=False)[0:5]

industry = ind.index
amount = ind.values
perIndustry = np.true_divide(amount,amount.sum())*100
for i in range(len(industry)):
    print(industry[i],format(perIndustry[i],'.2f'))
    
```

```python
# 3..
# Find out if cities play any role in receiving funding.
# Find top 10 Indian cities with most amount of fundings received. 
# Find out percentage of funding each city has got (among top 10 Indian cities only).
# Print the city and percentage with 2 decimal place after rounding off.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def handle_city(name):     #added this function to handle multiple city nams.....in case of multiple names...u need to take the first one
    return name.split('/')[0].strip()

data = pd.read_csv("Dataset/startup_funding.csv")
df1 = data.copy()

df1['CityLocation'].dropna(inplace=True)

df1['CityLocation'].replace("Delhi","New Delhi",inplace=True)
df1['CityLocation'].replace("bangalore","Bangalore",inplace=True)

df1['CityLocation'] = df1['CityLocation'].apply(handle_city)

df1['AmountInUSD'] = df1['AmountInUSD'].apply(lambda x :float(str(x).replace(",","")))
df1['AmountInUSD'] = pd.to_numeric(df1['AmountInUSD'])
start_fund = df1.groupby('CityLocation')['AmountInUSD'].sum().sort_values(ascending=False)[0:10]
city = start_fund.index
city_amount = start_fund.values
amount = np.true_divide(city_amount,city_amount.sum())*100
for i in range(len(city)):
    print(city[i],format(amount[i],'.2f'))
```

```python
import csv
from collections import Counter
import pandas as pd
import numpy as np

with open("Dataset/startup_funding.csv") as file_obj:
    reader = pd.read_csv(file_obj)
    
    reader['CityLocation'].dropna(inplace = True)
    
    df = reader[['CityLocation', 'InvestorsName']]
    df['CityLocation'].dropna(inplace = True)
    
    df['CityLocation'].replace("Delhi", "New Delhi", inplace = True)
    df['CityLocation'].replace("bangalore", "Bangalore", inplace = True)
    d= {}
    for index, row in df.iterrows():
        if(type(row['CityLocation']) == float or type(row['InvestorsName']) == float):
            continue;
        location = row['CityLocation']
        investor = row['InvestorsName'].split(',')
        location = location.strip().rstrip()
        investor = list(map(str.strip, investor))
        investor = list(map(str.rstrip, investor))
        investor = list(map(str.title, investor))
        if('/' in location):
            k = location.split('/')
            k = list(map(str.strip, k))
            k = list(map(str.rstrip, k))
            for i in range(2):
                for j in investor:
                        if(j == "Undisclosed Investors" or j == "") :
                            continue
                        if(j in d):
                            d[j] += 1
                        else :
                            d[j] = 1
                            
            for i in investor:
                if(i == "Undisclosed Investors" or i == ""):
                    continue
                if(i in d):
                    d[i] += 1
                else:
                    d[i] = 1
k = Counter(d) 
high = k.most_common(5)  
for i in high:
    print(i[0], i[1])
```

```python
with open("Dataset/startup_funding.csv") as file_obj:
    reader = pd.read_csv(file_obj)
    reader = reader[pd.notna(reader['InvestorsName'])]
    
    df = reader['InvestorsName']
    df.str.title()
    
    d = {}
    for row in df.iteritems():
        k = row[1].split(', ')
#         print(k)
        for i in k:
            if(i in d):
                d[i] += 1
            else :
                d[i] = 1

ans = -1
val = ""
for i in d:
    if(d[i] > ans):
        ans = d[i]
        val = i
print(val, ans)
```
